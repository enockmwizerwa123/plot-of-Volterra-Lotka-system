```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def RK4OdeSys(f,x0,t):
    H = t[1] - t[0]
    N = len(t)
    M = len(x0)
    x = np.zeros((N,M))
    x[0] = x0
    for k in range(0,N-1):
        k1 = f(x[k],t[k])
        k2 = f(x[k]+H*(k1/2),t[k]+H/2)
        k3 = f(x[k]+H*(k2/2),t[k]+H/2)
        k4 = f(x[k]+H*k3,t[k]+H)
        x[k+1] = x[k]+H/6*(k1+2*k2+2*k3+k4)  
    return x
```

```python
def Volterra_Lotka(X,t):
    B,R=X
    Bdt=B*(1-R)
    Rdt=R*(-1+B)
    Ans=np.array([Bdt,Rdt])
    return Ans
```

EULER METHOD

```python
def Eurler(VL,incond,t_steps,n):
    t=np.linspace(t_steps[0],t_steps[1],n)
    A=np.zeros((n,len(incond)))
    H=t[1]-t[0]
    A[0]=incond
    for j in range(1,n):
        A[j] = A[j-1]+H*VL(A[j-1],t[j-1])
    return A
incond=[2,1]
t_steps=[0,20]
n=100
t = np.linspace(t_steps[0],t_steps[1],n)
    
```

 RUNGE-KUTTA PLOT

```python
Runge_kutta=RK4OdeSys(Volterra_Lotka,incond,t)
plt.figure(figsize =(15,9))
plt.subplot(1,2,1)
plt.plot(t,Runge_kutta[:,0],label='R(t)',color ='m')
plt.plot(t,Runge_kutta[:,1],label='B(t)',color ='y')
plt.title(' B(t) And R(t) PLOT')
plt.xlabel('TIME')
plt.ylabel('B(t),R(t)')
plt.legend()
plt.subplot(1,2,2)
plt.plot(Runge_kutta[:,0],Runge_kutta[:,1],color ='k')
plt.xlabel('R(t)')
plt.ylabel('B(t)')
plt.title('Solution Trajectories')
plt.show()

```

EULER PLOT

```python
Euler=Eurler(Volterra_Lotka,incond,t_steps,n)
plt.figure(figsize =(10,6))
plt.subplot(1,2,1)
plt.plot(t,Euler[:,0],label='R(t)',color ='m')
plt.plot(t,Euler[:,1],label='B(t)',color ='y')
plt.title(' B(t) And R(t) Plot')
plt.xlabel('TIME')
plt.ylabel('B(t),R(t)')
plt.legend()
plt.subplot(1,2,2)
plt.plot(Euler[:,0],Euler[:,1],color ='b')
plt.xlabel('R(t)')
plt.ylabel('B(t)')
plt.title('Solution Trajectories')
plt.show()
```

the Euler method (also called forward Euler method) is a first-order numerical procedure for solving ordinary differential equations with a given initial value.When we change the initial conditions both the trajectory and plot for Euler's method tend to be smooth compared to the previous one i.e Euler's method is sensitive to the initial condition

Runge–Kutta method is an effective and widely used method for solving the initial-value problems of differential equations.In our case we can see that runge kutta method is the one which is giving accurate than euler method and also runge kutta method is the modiffication of euler method
